1. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{(y_{i} - (b + mx_{i}))^{2}}$$

&emsp;&emsp;де x,y - деякі задані вектори

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
x = np.random.normal(0, 5, 100)
y = 50 + 2 * x + np.random.normal(0, 2, len(x))
```

$$f' = \frac{-2}{n}\sum_{i=1}^{n}{(y_{i} - b - mx_{i})*x}$$
$$f' = \frac{-2}{n}\sum_{i=1}^{n}{(y_{i} - b - mx_{i})*1}$$

```python
def f(b, m):
    return np.mean((y-(b+m*x))**2)

def df(b, m):
    h = 1e-10
    df_b = (f(b + h, m) - f(b, m))/h
    df_m = (f(b, m + h) - f(b, m))/h
    return np.array([df_b, df_m])
```

```python
df(1, 1)
```

```python
w0 = np.zeros(2)
alpha = 0.01
epsilon = 0.00001

print(w0, f(w0[0], w0[1]))

while True:
    b = w0[0]
    m = w0[1]
    w = w0 - alpha * df(b, m)
    
    if np.sum(np.abs(w - w0)) <= epsilon:
        break
    else:
        w0 = w
            
    print(w, f(w[0], w[1]))
    

print('=' * 100)
print('Значення w, при якому f набуває мінімального значення: ', w)
print('Мінімум f: ', f(w[0], w[1]))
```

2. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{[(y_{i} - (b + mx_{i}))^{2}]} + m^2 + b^2$$

&emsp;&emsp;де x,y - вектори з 1-го завдання

```python
def f2(b, m):
    return np.mean((y-(b+m*x))**2)+m**2+b**2

def df2(b, m):
    h = 1e-10
    df_b = (f2(b + h, m) - f2(b, m))/h
    df_m = (f2(b, m + h) - f2(b, m))/h
    return np.array([df_b, df_m])
```

```python
w0 = np.ones(2)
alpha = 0.001
epsilon = 0.00001

print(w0, f2(w0[0], w0[1]))

while True:
    b = w0[0]
    m = w0[1]
    w = w0 - alpha * df2(b, m)
    
    if np.sum(np.abs(w - w0)) <= epsilon:
        break
    else:
        w0 = w
            
    print(w, f2(w[0], w[1]))
    

print('=' * 100)
print('Значення w, при якому f набуває мінімального значення: ', w)
print('Мінімум f: ', f2(w[0], w[1]))
```

3. Знайти значення а, для якого функція набуває мінімального значення

$$f = -\frac{1}{n}\sum_{i=1}^{n}{[y_{i}log(h(x_{i})) + (1 - y_{i})log(1 - h(x_{i}))]}$$

$$h = \frac{1} {1 + exp^{-ax}}$$

```python
np.random.seed(10)
num_observations = 100
x1 = np.random.normal(1, 1, 50)
x2 = np.random.normal(3, 1, 50)

x = np.concatenate([x1, x2])
y = np.concatenate([np.zeros_like(x1), np.ones_like(x2)])
```

```python
def f3(a):
    return np.mean(y*np.log((1 / (1 + np.exp((-a)*x)))) + (1 - y) * np.log(1 - (1 / (1 + np.exp((-a)*x)))))*(-1)

def df3(a): 
    j = 1e-10
    return (f3(a + j) - f3(a)) / j
```

```python
a0 = 0  # значення х, з якого ми почнемо пошук
alpha = 0.001  # крок алгоритма
epsilon = 0.00001  # критерій, якщо зміна в x менше "epsilon", то вважаємо, що рішення знайдено


while True:
    a = a0 - alpha * df3(a0)  # крок алгоритму, перехід до наступного значення х
    
    if abs(a - a0) <= epsilon:  # перевірка (тут, можливо, пошук варто припинити)
        break
    else:
        a0 = a  # оновлення попереднього значення
        
    print(a, f3(a))  # виведення поточного значення х та y
    
print('=' * 100)
print('Значення a, при якому f набуває мінімального значення: ', a)
print('Мінімум f: ', f3(a))
```
