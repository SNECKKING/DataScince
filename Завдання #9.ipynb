Згенегуємо два набори значень $x$ і $y$.

```python
import numpy as np

x = np.random.uniform(0, 100, 100)
y = 2 * x + 1 + np.random.randint(0, 25, 100)
```

```python
from matplotlib import pyplot as plt

plt.scatter(x, y)
```

### Завдання.

Знайти значення коефіцієнтів рівняння лінійної регресії $\beta_0$ та $\beta_1$:

1) за допомогою формул з підручника;

2) за допомогою моделі sklearn.linear_model.LinearRegression;

3) знайти передбачення для значення змінної $x$, що рівне 120. 

```python
#9.1
b1 = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x)**2)
b0 = np.mean(y) - b1 * np.mean(x)

print("B0: ", b0)
print("B1: ", b1)
plt.scatter(x, y)
plt.plot(x, b0+b1*x, c='red')
```

```python
#9.2
from sklearn.linear_model import LinearRegression
xr = x.reshape(-1, 1)
model = LinearRegression()
model.fit(xr, y)
```

```python
b1_2 = model.coef_
b0_2 = model.intercept_ 

print("B0: ", b0_2)
print("B1: ", b1_2)
plt.scatter(x, y)
plt.plot(xr, model.predict(xr), c='red')
print(model.score(xr, y))
```
